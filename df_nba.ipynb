```python
def get_na(df):
    qsna = df.shape[0] - df.isnull().sum(axis = 0) #Cantidad de NA
    qna = df.isnull().sum(axis = 0)
    ppna = round(100*(qna/df.shape[0]),2)
    aux = {'Datos sin Nas en qtd': qsna, 'Na en qtd': qna, 'Na en %': ppna}
    na = pd.DataFrame(data = aux)
    return na.sort_values(by = 'Na en qtd', ascending = False)
```

```python
# prompt: pasar a big query

import google.cloud.bigquery as bq

# Create a BigQuery client object.
client = bq.Client()

# Define the dataset and table names.
dataset_name = 'nba_stats'
table_name = 'games'

# Create a dataset reference.
dataset_ref = client.dataset(dataset_name)

# Create a table reference.
table_ref = dataset_ref.table(table_name)

# Load the parquet file into the table.
job_config = bq.LoadJobConfig(write_disposition='WRITE_TRUNCATE')
job = client.load_table_from_parquet('df_games.parquet', table_ref, job_config=job_config)
job.result()  # Waits for the job to complete.

# Define the dataset and table names for the other tables.
table_names = ['games_details', 'players', 'ranking', 'teams']

# Load the other parquet files into their respective tables.
for table_name in table_names:
  table_ref = dataset_ref.table(table_name)
  job_config = bq.LoadJobConfig(write_disposition='WRITE_TRUNCATE')
  job = client.load_table_from_parquet(f'{table_name}.parquet', table_ref, job_config=job_config)
  job.result()  # Waits for the job to complete.

print('Successfully loaded parquet files into BigQuery')
```
